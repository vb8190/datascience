## What is the true normal human body temperature? 

#### Background

The mean normal body temperature was held to be 37$^{\circ}$C or 98.6$^{\circ}$F for more than 120 years since it was first conceptualized and reported by Carl Wunderlich in a famous 1868 book. In 1992, this value was revised to 36.8$^{\circ}$C or 98.2$^{\circ}$F.

```python
import pandas as pd
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python

df = pd.read_csv('data/human_body_temperature.csv')
df
```

```python
#Question-1
temp = df['temperature']
scipy.stats.shapiro(temp)
```

```python
#Since the shapiro test gives us p-value = 0.233 that means we can reject NULL Hypothesis, 
#i.e. temperature variable is not normally distributed
```

```python
ig, axes = plt.subplots()

axes.hist(temp)
axes.set_title("Temperature Distribution")

#The distrbution looks kind of normal visually but shapiro test disproves this.
```

```python
#Question-2
scipy.stats.ttest_1samp(temp, 98.6)
#Since the p-value is too low, that means we can reject our null hypothesis that the sample mean is same 
#as population mean of 98.6, i.e. this sample did not come from population mean of 98.6

#We can not perform z-test because for that we need to know population standard deviation, which is not known.
```
